```python
df.isnull().sum()
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Gender'].isnull().sum()
```

```python
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
```

```python
df['Married'].isnull().sum()
```

```python
df['Dependents'] = (
    df['Dependents']
    .replace('3+', 3)
    .astype(float)
)
```

```python
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].median()).astype(int)
```

```python
df['Dependents'].isnull().sum()
```

```python
df['Education']=df['Education'].fillna(df['Education'].mode()[0])
```

```python
df['Education'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
df['Self_Employed'].isnull().sum()
```

```python
df['CoapplicantIncome']=df['CoapplicantIncome'].fillna(df['CoapplicantIncome'].median())
```

```python
df['CoapplicantIncome'].isnull().sum()
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].median())
```

```python
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])
```

```python
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df['Property_Area']=df['Property_Area'].fillna(df['Property_Area'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.sample(3)
```

```python
df['Loan_ID']=df['Loan_ID'].astype(str)
```

```python
cat_cols=['Gender','Married','Education','Self_Employed','Property_Area']
df[cat_cols] = df[cat_cols].astype('category')
```

```python
int_cols=['Dependents','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']
df[int_cols]=df[int_cols].astype(int)
```
